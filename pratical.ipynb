```python
#1d array 
import numpy as np
a=np.array([1,2,3,4,5])
print(a)
```

```python
#shape of array 
print("shape",a.shape)
```

```python
print("dimension",a.ndim)
```

```python
print("size",a.size)
```

```python
print("data type",a.dtype)
```

```python
print("size of item",a.itemsize)
```

```python
# create a 2d array with oparation 
b=np.array([[1,2,3],[4,5,6]])
c=np.array([[8,9,10],[11,12,13]])
print(b)
print(c)
```

```python
# add a 2 array 
add=np.add(b,c)
print("add:",add)
```

```python
# subtract with 2 array
subtract=np.subtract(b,c)
print("subtract:",subtract)
```

```python
# divide 
divide=np.divide(b,c)
print("divide:",divide)
```

```python
# multiply 
multiply=np.multiply(b,c)
print("multiply:",multiply)
```

```python
mod=np.mod(b,c)
print("mod two array:",mod)
```

```python
remainder=np.remainder(b,c)
print("remainder:",remainder)
```

```python
power=np.power(b,c)
print("power:",power)
```

```python
absolute=np.absolute(b)
print("absolute:",absolute)
```

```python
print("sum : ",np.sum(b))
print("sqrt : ",np.sqrt(b))
```
